```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
```

```python
# Remote URL to data
url = "https://docs.google.com/spreadsheets/d/e/2PACX-1vR_e8C18AejPhQs1YeE9YWo8HtsUymYT3nBiqTi7-2M0LUncBStzTWD3PdDyYgZFcZ5q5T1NRM77NHV/pub?output=csv"
# Load the data
df = pd.read_csv(url)
# Verify it loaded
df.head()
```

```python
# Get the target column's mean
mean_1 = df["VitD_levels"].mean()
# Get the target column's median
median_1 = df["VitD_levels"].median()
# Get the target column's +1 STD
std_plus_one_1 = mean_1 + df["VitD_levels"].std()
# Get the target column's -1 STD
std_minus_one_1 = mean_1 - df["VitD_levels"].std()
```

```python
# Define figure and axes
fig, ax = plt.subplots(figsize=(10, 6))
sns.histplot(data=df, x="VitD_levels", stat="probability", kde=True)
# Set the title
ax.set_title("Vitamin D Levels")
# Draw the mean line
ax.axvline(mean_1, 
           color="red",
           label=f"Mean = {round(mean_1, 2)}",
           linewidth=4)
# Draw the median line
ax.axvline(median_1, 
           color="green",
           label=f"Median = {round(median_1, 2)}",
           linewidth=2,
           ls="--")
# Draw the -1 STD line
ax.axvline(std_minus_one_1, 
           color="black",
           label=f"-1 STD = {round(std_minus_one_1, 2)}",
           linewidth=3)
# Draw the +1 STD lin
ax.axvline(std_plus_one_1, 
           color="black",
           label=f"+1 STD = {round(std_plus_one_1, 2)}",
           linewidth=3)
# Highlight the range between -1 STD and +1 STD
ax.axvspan(std_minus_one_1, std_plus_one_1, color="yellow", zorder=0)
# Supply a legend
ax.legend();
```

```python
# Calculate kurtosis
kurtosis = stats.kurtosis(df["VitD_levels"] , fisher=False)
kurtosis
```

1. Is it Discrete or Continuous?
> * This is a continuous distribution. 
2. Does it have a skew? If so, which direction (+/-)
> * This does not have a significant skew. 
3. What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
> * The distribution above displays Mesokurtic Kurtosis.
